```python
covid_data.drop(['Sno','ConfirmedIndianNational','ConfirmedForeignNational','Time'],inplace = True,axis =1)
```

```python
covid_data.head()
```

```python
covid_data['Date'] = pd.to_datetime(covid_data['Date'],format = '%Y-%m-%d')
```

```python
covid_data.head()
```

```python
covid_data.info()
```

```python
covid_data['Active_Cases'] = covid_data['Confirmed'] -(covid_data['Cured']+covid_data['Deaths'])
covid_data.head()
```

```python
statewise = pd.pivot_table(covid_data,values = ["Confirmed",'Deaths','Cured'],index ='State/UnionTerritory',aggfunc = max )
```

```python
statewise
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise
```

```python
statewise = statewise.sort_values(by = 'Confirmed',ascending = False)
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
top_10_active_cases =covid_data.groupby('State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by =['Active_Cases'],ascending = False).reset_index()
```

```python
top_10_active_cases =covid_data.groupby('State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by =['Active_Cases'],ascending = False).reset_index()
fig = plt.figure(figsize=(14,8))
plt.title("10 states with most active cases in india",size=20)
sns.barplot(data = top_10_active_cases.iloc[:10],y='Active_Cases',x='State/UnionTerritory')
plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()
```

```python
top_10_active_cases =covid_data.groupby('State/UnionTerritory').max()[['Deaths','Date']].sort_values(by =['Deaths'],ascending = False).reset_index()
fig = plt.figure(figsize=(14,8))
plt.title("10 states with most deaths in india",size=20)
sns.barplot(data = top_10_active_cases.iloc[:10],y='Deaths',x='State/UnionTerritory',linewidth =2 ,edgecolor = 'black' )
plt.xlabel('States')
plt.ylabel('Total Deaths')
plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
sns.lineplot(data = covid_data[covid_data['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Delhi','Tamil Nadu','Uttar Pradesh'])
                              ,x='Date',y='Active_Cases',hue = 'State/UnionTerritory'])
plt.show()
                              
```

```python
male = vaccine_data['Male (Doses Administered)'].sum()
female = vaccine_data['Female (Doses Administered)'].sum()
plt.title("Vaccination")
plt.pie([male,female],labels =['Male','Female'])
plt.show()
```

```python
max_vac = vaccine_data.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
max_vac= max_vac.sort_values('Total Individuals Vaccinated',ascending = False)[:5]
max_vac
```

```python

fig = plt.figure(figsize=(14,8))
plt.title("Top 5 vaccinated states",size=20)
sns.barplot(data = max_vac.iloc[:10],y=max_vac.Total Individuals Vaccinated,x=max_vac.index,linewidth =2 ,edgecolor = 'black' )
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
